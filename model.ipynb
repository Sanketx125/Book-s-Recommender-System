# Popularity Based Recommender:

* by using rating of books we give suggrestions of top 50 books which can give more than 250 user to the books

* 

```python
ratings_with_name = rating.merge(books, on= 'ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns= {'Book-Rating': 'num_ratings'}, inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns= {'Book-Rating': 'avg_ratings'}, inplace=True)
```

```python
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on= 'Book-Title')
```

```python
popularity_df
```

```python
popularity_df = popularity_df[ popularity_df['num_ratings'] >= 250 ].sort_values('avg_ratings', ascending= False).head(50)
```

```python
popularity_df
```

```python
popularity_df = popularity_df.merge(books, on= 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_ratings']]
```

```python
popularity_df
```

## Colaborative Filtering Based Recommender System:

```python
ratings_with_name.head(3)
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filter_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filter_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50

famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

# applying pivot table

```python
pt = final_ratings.pivot_table(index= 'Book-Title', columns= 'User-ID', values= 'Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

# importing Cosin-Similarity:

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

# recommend function:

```python
def recommend(book_name):

    # fetching index by using book name:
    index = np.where( pt.index == book_name )[0][0]

    # finding distances by using similarity score:
    similar_items = sorted( list( enumerate( similarity_score[index] )), key = lambda x: x[1], reverse= True)[ 1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```
